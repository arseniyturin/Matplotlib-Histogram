# How To Make Awesome Histogram

This is quick tutorial on how to customize matplotlib histogram to impress your co-workers or the boss

## Importing Libraries

Let's start with importing necessery libraries

We will need `matplotlib` for plotting and `numpy` for generating test data

```python
# Importing libraries
import matplotlib.pyplot as plt # keyword 'as' let us get access to entire library through 'plt' alias
import numpy as np # numpy is a powerful library for scientific computing. We will use it to generate normal distribution
```

## Generating Normal Distribution

```python
# x will contain an array of data
np.random.seed(1)
x = np.random.normal(0, 1, 6000)
```

```python
# Simplest way to plot a histogram
plt.hist(x)
```

## How to Customize?

Matplotlib has a lot of methods and options to change the appearance of your plot.

### Styles
Matplotlib has built-in styles and colormaps to choose from. Style is basically how background of your plot would look like. To access all available styles we will use `plt.style.available` method.
### Colormaps
Colormaps on the other hand are just ranges of colors your plot would use for coloring shapes (bars, dots, lines, etc..)

For additional information on colormaps please check <a href="https://matplotlib.org/tutorials/colors/colormaps.html#sphx-glr-tutorials-colors-colormaps-py" target="_blank">official documentation</a>

```python
# Styles available
print( plt.style.available[:20])
```

```python
# Colormaps available
print( plt.colormaps()[:20] ) # Notice we are showing only first 20 available options to keep it clean
```

I personally like `seaborn-whitegrid` style

```python
plt.figure(figsize=(10,5)) # Figure size
plt.style.use('seaborn-whitegrid') # Selecting whitegrid style

# Building a histogram with custom color and number of bins
plt.hist(x, bins=90, facecolor = '#2ab0ff', edgecolor='#169acf', linewidth=0.5)

# Adding title and labels
plt.title('Normal Distribution')
plt.xlabel('Distribution')
plt.ylabel('Frequency')
plt.show()
```

## Make It Shine!

```python
plt.figure(figsize=(10,5)) # Plot size
plt.style.use('seaborn-whitegrid') # Style

# Building a histogram
n, bins, patches = plt.hist(x, bins=90, facecolor='#2ab0ff', edgecolor='#e0e0e0', linewidth=0.5, alpha=0.7)

# Changing properties of each bin
n = n.astype('int') # n should be integer, otherwise it wouldn't work properly

# Accessing each bin, changing color according to height
for i in range(len(patches)):
    patches[i].set_facecolor(plt.cm.viridis(n[i]/max(n)))
    
# Pick one bin
patches[47].set_fc('red') # Color
patches[47].set_alpha(1) # Opacity

# Making annotation
plt.annotate('Something important here!', xy=(0.57, 175), xytext=(2, 130), fontsize=10, arrowprops={'width':0.4,'headwidth':7,'color':'#333333'})

# Adding title and labels with custom font sizes
plt.title('Normal Distribution', fontsize=12)
plt.xlabel('Distribution', fontsize=10)
plt.ylabel('Frequency', fontsize=10)
plt.show()
```
